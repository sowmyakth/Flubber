```python
import daft
```

```python
# Instantiate a PGM.
pgm = daft.PGM([4.3, 2.05], origin=[0.3, 0.3], grid_unit=4., node_unit=1.6, observed_style="inner")
#pgm = daft.PGM([4.3, 4.0], origin=[0.3, 0.3], grid_unit=2.6, node_unit=1.3, observed_style="inner")
# Hierarchical parameters.
pgm.add_node(daft.Node("Sigma,sigma","$\{\mu^g,\sigma^g,w^g\}$", 0.8, 2,scale=1.4))
pgm.add_node(daft.Node("g", r"$g$", 1.5, 2,scale=1.4))
pgm.add_node(daft.Node("Ng", r"$N_g$", 0.45, 1.6, fixed=True,offset=(-4,-20)))

# Latent variable:
pgm.add_node(daft.Node("sigmaobs", r"$\sigma_{k}^{\rm obs}$", 2, 0.5,fixed=True,offset=(13,0)))
pgm.add_node(daft.Node("e", r"$e_k}$", 1, 1,scale=1.4))
pgm.add_node(daft.Node("elensed", r"$e_k^{lensed}$", 1.5, 1, fixed=True,offset=(3,-25)))





# Data:
pgm.add_node(daft.Node("eobs", r"$e_{k}^{\rm obs}$", 2, 1,observed=True,scale=1.4))

# Add in the edges.
pgm.add_edge("Ng", "Sigma,sigma")
pgm.add_edge("Sigma,sigma","e")
pgm.add_edge("g", "elensed")
pgm.add_edge("e", "elensed")
pgm.add_edge("elensed", "eobs")
pgm.add_edge("sigmaobs", "eobs")



# And a plate.
pgm.add_plate(daft.Plate([0.5, 0.5, 2, 1], label=r"galaxies $k=1, \cdots, K$",
    shift=-0.1))

# Render and save.
pgm.render()
pgm.figure.savefig("flubber_PGM.png", dpi=300)
```
